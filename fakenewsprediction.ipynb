```python
import pandas as pd
import numpy as np
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

import nltk
nltk.download('stopwords')

# Load data
news_dataset = pd.read_csv('/content/news.csv.zip', encoding='latin1')  # or 'ISO-8859-1'


# Fix text + title merge
news_dataset['content'] = news_dataset['title'] + " " + news_dataset['text']

# Map labels
news_dataset['label'] = news_dataset['label'].map({'FAKE': 0, 'REAL': 1})

# Clean text
port_stem = PorterStemmer()
def stemming(content):
    content = re.sub('[^a-zA-Z]', ' ', content)
    content = content.lower().split()
    content = [port_stem.stem(word) for word in content if word not in stopwords.words('english')]
    return ' '.join(content)

news_dataset['content'] = news_dataset['content'].apply(stemming)

# TF-IDF vectorization
vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X = vectorizer.fit_transform(news_dataset['content'].values)
Y = news_dataset['label'].values

# Train-Test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)

# Model
model = LogisticRegression()
model.fit(X_train, Y_train)

# Accuracy
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print("Training Accuracy:", accuracy_score(Y_train, train_pred))
print("Testing Accuracy:", accuracy_score(Y_test, test_pred))
```
